```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('combined-files_droped_null.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['warehouse'] = df['warehouse'].str.replace('دارد','True')
df['warehouse'] = df['warehouse'].str.replace('ندارد','True')
```

```python
df['parking'] = df['parking'].str.replace('دارد','True')
df['parking'] = df['parking'].str.replace('ندارد','True')
```

```python
df['elevator'] = df['elevator'].str.replace('دارد','True')
df['elevator'] = df['elevator'].str.replace('ندارد','True')
```

```python
df.head()
```

```python
df['type'].unique()
```

```python
df['type'] = df['type'].str.replace('آپارتمان','Apartment')
df['type'] = df['type'].str.replace('ویلا','Vilaa')
df['type'] = df['type'].str.replace('خانه و کلنگی','House')
df.head()
```

```python
df['type'].unique()
```

```python
df['age'] = df['age'].str.replace('سال', '')
df.head()
```

```python
df['age'].unique()
```

```python
df['age'] = df['age'].str.replace('نوساز', '0')
#more than 30 ==> 30+
df['age'] = df['age'].str.replace('بیشتر از 30 ', '30+')
df['age'].unique()
```

```python
df.head()
```

```python
df['rooms'].unique()
```

```python
# more than 5 ==> 5+
df['rooms'] = df['rooms'].str.replace('5 به بالا', '5+')
df.head()
```

```python
df['Address'] = df['Address'].str.replace('تهران، تهران، ', '')
df
```

```python
df['roof_garden'] = df.apply(lambda x: 'روف گاردن' in x.Descriprion, axis=1)
df['teracce'] = df.apply(lambda x: 'تراس' or 'بالکن' in x.Descriprion, axis=1)
df['teracce'] = df.apply(lambda x: ('استخر' or 'سونا/جکوزی' or 'جکوزی' or 'سونا') in x.Descriprion, axis=1)
df['party_room'] = df.apply(lambda x: ' سالن اجتماعات' in x.Descriprion, axis=1)
df
```
