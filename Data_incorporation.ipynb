```python
def neighbourhood_cleaner(n):
    n = n.replace("تهران"  , "")
    n = n.replace("," , "")
    n = n.replace("،" , "")
    n = n.replace(" " , "")
    n = n.split("-")[0]
    # n.strip()
    # n = n.split()[0] if (len(n.split())>=1 ) else ""
    n = n.strip()

    return n
    
```
